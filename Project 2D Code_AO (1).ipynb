## Scientific Question: Does the maintenance of genomic stability in great white sharks occur mainly due to their large number of histone genes (compared to other species and because of their much larger genome requiring efficient DNA packaging), or from the positive selection and accumulation of specific histone genes involved in acetylation and methylation?

Histones are proteins that provide structural support for chromosomes (as the DNA is wound tightly around the hsitones to create nucleosomes). Another important function for histone modifications is  DNA damage response, where due to their large genome and the associated requirements for packing DNA efficiently, great white sharks have a high number of core histone proteins present in the genome (notably with a large amount of H2A histone genes). Phosphorylation of H2AX is associated with the recruitment of DNA repair proteins, working to ensure genomic stability along with other notable histones with other functions, such as H3K56. Histone variants play a role in the DNA damage signaling and repair pathway choice, especially the H2A histone variants which are abundant in sharks. When there is less acetylation of histones then the created DNA is more likely to be damaged, with the DNA damage repair then wanting to work to preserve acetylation. However, less histone acetylation occurrs when there is less DNA damage, so acetylation is often retained at histone sites near DNA damage in places it is normally de-acetylated (when no damage is detected by the DNA damage checkpoint proteins).

```python
#there are a pair of input functions, read and parse. Read used when there is one object, and parse used when there are lots of objects in a BLAST record

# NCBIXML.parse() handles many objects and returns and allows for the BLAST records to be searched
E_VALUE_THRESH = 0.05
from Bio.Blast import NCBIXML
blast_records = NCBIXML.parse(open("H2AX_GWS_BLASTXresults.xml", 'r'))


for record in NCBIXML.parse(open("H2AX_GWS_BLASTXresults.xml")): 
     if record.alignments: 
        print("\n") 
        print("query: %s" % record.query[:100]) 
        for align in record.alignments:
            for hsp in align.hsps:   
                if hsp.expect < E_VALUE_THRESH:
                    print("match: %s " % align.title)
                    print('****Alignment****')
                    print('sequence:', align.title)
                    print('length:', align.length)
                    print('identity', hsp.identities)
                    print('% identity', ((hsp.identities) / (align.length)))
                    print('e value:', hsp.expect)
                    print(hsp.query[0:85] + '...')
                    print(hsp.match[0:85] + '...')
                    print(hsp.sbjct[0:85] + '...')
```

MultipleSeqAlignment takes the .aln file and the genes that are within the dataset and the TreeConstruction can be performed, with the resulting distance matrix being created that allows for results to be plotted in both the phylogenetic tree and the heatmap.

Phylogenetic tree

```python
# Open the alignment file as a MultipleSeqAlignment object 
with open("muscleresultsh2ax.aln","r") as aln: 
    alignment = AlignIO.read(aln,"clustal")
print(type(alignment))
```

```python
# Open and initiate the Distance Calculator using the Identity model  
calculator = DistanceCalculator('identity')
```

```python
# Writing the Distance Matrix, where distance_matrix is a global variable
distance_matrix = calculator.get_distance(alignment)
print(distance_matrix)
```

```python
#here is the code to download the distance matrix as a .txt file to be manipulated as a .csv for further analysis using excel
file1 = open("distance_matrix.txt", "w") 
file1.write("%s = %s\n" %("distance_matrix", distance_matrix))
file1.close()
```

```python
# Opening and initiating the Tree Constructor 
constructor = DistanceTreeConstructor(calculator)
```

```python
# Building the tree, with H2AX_tree being a global variable
H2AX_tree = constructor.build_tree(alignment)
H2AX_tree.rooted = True
print(H2AX_tree)
```

```python
# Saving the tree to a new .xml file 
Phylo.write(H2AX_tree, "H2AX_tree.xml", "phyloxml")
```

```python
# Making a phylogenetic tree using features from matplotlib 
# https://www.youtube.com/watch?v=wBdz3vFQ4Ks

fig = plt.figure(figsize=(13, 5), dpi=100) # creating figure & set the size 
matplotlib.rc('font', size=6)               
matplotlib.rc('xtick', labelsize=10)       # fontsize of the x labels
matplotlib.rc('ytick', labelsize=10)       # fontsize of the y labels
H2AX_tree.ladderize()
axes = fig.add_subplot(1, 1, 1)
Phylo.draw(H2AX_tree, axes=axes)
fig.savefig("H2AX_cladogram")
```
